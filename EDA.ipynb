we can see that most of the columns have uniform distribution, except column 6, and column 8 that stores binary values. So, let's check if any of these features correlate, to avoid multicollinearity

```python
corr_m = data.corr()
```

```python
corr_m
```

```python
high_corr_pairs = []

for i in range(corr_m.shape[0]):
    for j in range(i + 1, corr_m.shape[1]):
        if abs(corr_m.iloc[i, j]) > 0.8:
            corr_value = corr_m.iloc[i, j]
            high_corr_pairs.append((corr_m.columns[i], corr_m.columns[j], corr_value))

print(high_corr_pairs)
```

So, we see that no features highly correlate with target, so we cannot represent target as a linear combination of some columns. But there is a high correlation between col 6 and 8. Let's check it one more time with another correlation type. Because it also seems, that column 8 values show whether values from col 6 are positive or negative.

```python
filtered_data = data[data['6'] > 0]
filtered_data[filtered_data['8'] != 1]
```

for the above assumption, I have checked if there are rows, where values from column 6 are positive, and values from column 8 are not equal to 1, and as a result there are no such values.

```python
spearman_corr = data.corr(method='spearman')
```

```python
high_corr_pairs1 = []
for i in range(spearman_corr.shape[0]):
    for j in range(i + 1, spearman_corr.shape[1]):
        if abs(spearman_corr.iloc[i, j]) > 0.8:
            corr_value = spearman_corr.iloc[i, j]
            high_corr_pairs1.append((spearman_corr.columns[i], spearman_corr.columns[j], corr_value))

print(high_corr_pairs1)
```

I have tried spearman correlation as well, that also works with non-linearity, and it also shows hight correlation between columns 6 and 8. So, to avoid multicollinearity or possible overfit, It would be better to drop column 8, since it doesn't store any unique information.

```python
data.drop(columns=['8'], inplace=True)
```
